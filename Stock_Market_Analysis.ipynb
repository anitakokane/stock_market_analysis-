# **Stock Market Analysis of Tesla(TSLA)**

This project analyzes Tesla’s stock data using Python libraries such as Pandas,NumPy, Matplotlib, and Seaborn. The goal is to understand pricing trends, stock volatility, daily returns, correlations, and overall market behavior.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Importing all essential Python libraries for analysis and visualization.

```python
df=pd.read_csv("/content/stock_data_tysla.csv")
df
```

The dataset contains historical stock data for Tesla (TSLA).

Columns include:

Date: Trading date

Open: Opening price

High: Highest price of the day

Low: Lowest price of the day

Close: Closing price

Adj Close: Adjusted closing price

Volume: Number of shares traded

```python
df.info()   # show basic Information about data
```

```python
import numpy as np


columns_to_modify = ['Open', 'High', 'Low', 'Close']


for col in columns_to_modify:

    num_nulls = int(len(df) * 0.1)
    null_indices = np.random.choice(df.index, num_nulls, replace=False)
    df.loc[null_indices, col] = np.nan

print("DataFrame with introduced null values:")
display(df.head())
display(df.info())
```

```python
df.set_index('Date', inplace=True)  #Making date column as index
```

```python
# Calculate the average closing price for each week by resampling the daily data into weekly groups.
# "W" means weekly frequency, and mean() gives the average Close price for every week.
weekly_avg = df['Close'].resample("W").mean()
```

```python
weekly_avg
```

```python
# Calculate the average closing price for each Month by resampling the daily data into Monthly groups.
# "M" means monthly frequency, and mean() gives the average Close price for every Month.
monthly_volume = df['Volume'].resample("M").sum()
monthly_volume
```

```python
# Calculate the daily percentage return of the stock.
# pct_change() finds how much the closing price changed compared to the previous day.
df["daily_return"] = df["Close"].pct_change()
df
```

```python
# Plotting the closing price over time to visualize how the stock price has changed.
# This line plot helps identify overall trends, upward/downward movements, and periods of volatility.
#line plot
plt.figure(figsize=(10,6))
df['Close'].plot()
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Time Series Plot of Close Price')
plt.tight_layout()
plt.grid()
plt.show()
```

This line plot shows how Tesla’s closing price changed over the entire dataset period.

Observations:

✔ The stock shows initial fluctuations, indicating high volatility in early months.

✔ After mid-2023, the price movement becomes more stable and gradually increases.

✔ From mid-2024 onwards, the closing price begins a strong upward trend, rising sharply.

✔ Sudden drops in the line indicate days with missing market activity or large negative returns.

Overall, the graph shows that Tesla’s stock experienced volatility initially but later moved into a clear upward trend, suggesting stronger market performance toward the end of the dataset.

```python
# Plotting the Open and Close prices together.
# This helps compare how the stock starts the day (Open) vs. how it ends the day (Close).
plt.figure(figsize=(10,6))
plt.plot(df['Open'], label='Open')
plt.plot(df['Close'], label='Close')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Plot of Open and Close Prices')
plt.legend()
plt.grid()
plt.show()
```

This graph compares the opening price and closing price of the stock on each day.

✔ Open and Close prices follow almost the same pattern.

✔ In the beginning, the prices go up and down quickly (high volatility).

✔ Later, both Open and Close prices show a clear upward trend.

✔ The Close price is sometimes higher or lower than the Open price, which shows daily gains or losses.

The stock became stable later and started increasing strongly, and Open/Close usually move together.

```python

plt.figure(figsize=(10,6))
plt.plot(df['High'], label='High')
plt.plot(df['Low'], label='low')
plt.xlabel('date')
plt.ylabel('Price')
plt.title('Plot of high and low Prices')
plt.grid(linestyle='--')
plt.legend()
plt.show()
```

This line plot shows how Tesla’s highest (High) and lowest (Low) prices changed over the entire time period in the dataset.

The High line shows the maximum price Tesla reached each day.

The Low line shows the minimum price Tesla touched each day.

The gap between the two lines represents the daily volatility of the stock.


✔ Early months show big fluctuations, meaning the stock was highly volatile in the beginning.
✔ The gap between the High and Low prices varies a lot early on, which means the price moved up and down very quickly.

✔ From late 2023 to mid-2024, the High and Low lines move smoothly and steadily upward, showing stable growth with less volatility.

✔ From late-2024 onwards, Tesla experiences a strong upward trend, with both High and Low prices rising sharply.

✔ Some sudden vertical lines appear—these are due to missing dates or gaps in the dataset, not market behavior.

✔ Overall, the graph shows Tesla’s stock becoming more stable over time and then accelerating into strong growth toward the end.

```python
plt.figure()
df['Close'].rolling(window=20).mean().plot(label='20-Day MA')
df['Close'].rolling(window=50).mean().plot(label='50-Day MA')

plt.title("Moving Averages (20 & 50 Days)")
plt.xlabel("Date")
plt.ylabel("Moving Average Price")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure()
plt.scatter(df['Volume'], df['Close'])
plt.title("Volume vs Close Price")
plt.xlabel("Volume")
plt.ylabel("Close Price")
plt.tight_layout()
plt.show()
```

Summary: The scatter plot shows a strong, non-linear positive correlation between the stock's Volume and its Close Price.

Result:

Low Volume: Prices are highly volatile (a wide range of Close Prices).

High Volume: The price forms a tight, upward-sloping curve, indicating that higher trading volume is strongly associated with higher stock prices.

```python
plt.figure()
df['daily_return'].plot(kind='hist', bins=30)
plt.title("Distribution of Daily Returns")
plt.xlabel("Daily Returns")
plt.tight_layout()
plt.show()
```

The distribution is leptokurtic (peaked) and centered around zero. This is typical for stock returns, showing that most days experience very small price changes.

✔️ Result

Stability: The massive central peak at 0.0 confirms that the stock's most common daily return is a near-zero change.

Risk: The "fat tails" (small bars extending far to the left and right) indicate a higher-than-normal probability of extreme positive or negative movements (large gains or losses). The stock is generally stable but subject to sudden, significant jumps in volatility.

```python
plt.figure()
plt.boxplot(df['Close'])
plt.title("Box Plot of Close Price")
plt.ylabel("Close Price")
plt.tight_layout()
plt.show()
```

✔ The central box is narrow for later dates, indicating reduced price variability and more stable trends.

✔ The median line is positioned toward the upper half, showing that closing prices tended to move upward.

✔ Several data points fall outside the whiskers, representing outliers — these are days with unusually high spikes or sudden price drops.

✔ The overall upward shift of the box suggests consistent long-term price growth.

Overall Insight

The boxplot confirms that the stock became more stable over time and followed a rising trend with occasional volatility spikes.

```python
plt.figure(figsize=(8,5))
sns.heatmap(df[['Open','High','Low','Close','Volume']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```


The heatmap visualizes how strongly different stock variables (Open, High, Low, Close, Volume) are correlated.

Observations

✔ Open, High, Low, and Close show very strong positive correlation, meaning they move together consistently.

✔ Close price has the highest correlation with High and Low — expected in stock price patterns.

✔ Volume has a weaker correlation with price features, indicating that trading volume does not directly drive price movement in this dataset.

✔ No negative correlations are observed among price features.


Price-based variables behave closely together, while Volume acts more independently, suggesting price trends are not solely volume-driven.





```python
df["price_change"] = df["Close"] - df["Open"]  #creating a new column to check daily changing price
df
```

```python
plt.plot(df["price_change"])
plt.xlabel("Date")
plt.ylabel("Price Change")
plt.title("Price Change Over Time")
plt.show()
```

The line plot tracks the daily Price Change (Y-axis) against the Date (X-axis).

✔️ Summary
The graph shows a highly erratic time series with two major periods of extreme volatility separated by a long period of relative stability.

✔️ Result
Initial and Final Volatility: The stock experienced large, sharp swings (both positive and negative changes) at the very beginning (Early 2023) and very end (Late 2024/Early 2025) of the dataset.

Mid-Term Stability: In between these two extreme episodes, the price change line stayed consistently near the zero mark (Mid 2023 to Late 2024), indicating a phase of low daily price movement and stability.
